```python
import numpy as np
import pandas as pd
from  matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm['Total']= sm[['English','Hindi'	,'Maths','Science','SST']].sum(axis=1)
sm
```

```python
sm['Percentage']= sm[['Total']]/5
sm
```

```python
#calculate the grades

def calc_grade(p):
    if p<=100 and p>=80:
        return 'A'
    elif p<80 and p>=60:
        return'B'
    elif p<60 and p>=50:
        return'C'
    elif p<50 and p>=40:
        return'D'
    elif p<40 and p>=30:
        return'E'
    elif p<30 and p>=0:
        return'F'
    else:
        return'xx'


sm['Grades']=sm['Percentage'].apply(calc_grade)
sm
```

```python


temp=sm.head()
plt.figure(figsize=(10,6))
for sub in ['English','Hindi','Maths','Science','SST']:
  plt.plot(temp['Name'],temp[sub],marker='o',label=sub)
  plt.title("Marks in different subjects")
  plt.xlabel("Name        ")
  plt.ylabel("Marks")
  plt.legend()
  plt.grid(True)
  plt.show()
```
